DBSCAN

```python
from numpy import where
from sklearn.datasets import make_classification, make_moons # Import make_moons
from sklearn.cluster import DBSCAN
import numpy as np

X, _ = make_moons(noise=0.05, random_state=0, n_samples=1000)

# Assuming pyplot was intended to be used, import it
import matplotlib.pyplot as plt

plt.scatter(X[:, 0], X[:, 1])
plt.show()
```

```python
from sklearn.cluster import DBSCAN

model = DBSCAN(eps=0.2, min_samples=9)
yhat = model.fit_predict(X)

clusters = np.unique(yhat)
```

```python
for cluster in clusters:
  row_ix= where(yhat == cluster)
  plt.scatter(X[row_ix, 0], X[row_ix, 1])
plt.show()
```

K-Means

```python
from sklearn.cluster import KMeans
from sklearn.datasets import make_classification

model= KMeans(n_clusters=3)
model.fit(X)
yhat= model.predict(X)

cluster=np.unique(yhat)

for cluster in clusters:
  row_ix= where(yhat == cluster)
  plt.scatter(X[row_ix, 0], X[row_ix, 1])
plt.show()
```
