```python
from matplotlib import pyplot as plt
import cv2
import numpy as np
```

```python
filename  = '/content/drive/MyDrive/photo-1486365227551-f3f90034a57c.jfif'

img=cv2.imread(filename)
plt.imshow(img)
```

```python
#interpolation

(height,width)= img.shape[:2]
res= cv2.resize(img,(int(width/2),int(height/2)),interpolation=cv2.INTER_CUBIC)

plt.imshow(res)
```

```python
# rotation

rows, cols= img.shape[:2]

M=cv2.getRotationMatrix2D((cols/2,rows/2),45,1)

res=cv2.warpAffine(img,M,(cols,rows))

plt.imshow(res)
```

```python
#shifting

M =np.float32([[1,0,80],[0,1,30]]) #[vertical],[horizontal]
rows,cols = img.shape[:2]

res=cv2.warpAffine(img,M,(cols,rows))

plt.imshow(res)
```

```python
# line detection

res = cv2.Canny(img,100,200)
plt.imshow(res)
```

```python
# blur - kernal;

kernel = np.ones((5,5),np.float32)/25
res = cv2.filter2D(img,-1,kernel)
plt.imshow(res)
```

```python
#constract or brightness

res= cv2.addWeighted(img,5, np.zeros(img.shape,img.dtype),0,0)
plt.imshow(res)
```

```python
#convert BW

res= cv2.cvtColor(img,cv2.COLOR_BGR2GRAY) # Changed cv2.color_BGR2GRAY to cv2.COLOR_BGR2GRAY
plt.imshow(res)
```
