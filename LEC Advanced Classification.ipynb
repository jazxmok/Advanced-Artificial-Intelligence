Import necassary library

```python
import pandas as pd
```

import and display the dataset. find other dataset that has missing value and imbalanced class (target)

```python
df =pd.read_csv('https://github.com/YBI-Foundation/Dataset/raw/refs/heads/main/Diabetes%20Missing%20Data.csv')
```

```python
df.head()
```

```python
df.shape
```

Check total missing value

```python
df.isnull().sum()
```

Method 1: Simply drop row of missing value

```python
df_drop_na = df.dropna()
```

```python
df_drop_na.isnull().sum()
```

```python
df_drop_na.shape
```

```python
df.shape
```

do classification

```python
X = df_drop_na.iloc[:,0:-1] #input
y = df_drop_na.iloc[:,-1] #target
```

```python
#divide the data into training and testing

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3)
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, accuracy_score,confusion_matrix

result_acc = accuracy_score(y_test,y_pred)
print(f'Accuracy: {result_acc}')

result_conf = confusion_matrix(y_test,y_pred)
print(f'Confusion Matrix: {result_conf}')
```
